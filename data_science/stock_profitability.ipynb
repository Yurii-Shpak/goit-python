```python
%matplotlib inline
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

shares = pd.DataFrame({'return_a': [25, -10, 10, 5, 35, 13], 
                       'return_b': [0, 15, -5, 5, 20, 25],
                       'return_c': [10, 25, -15, -5, -5, 15]
                      })
ax1 = sns.regplot(x="return_a", y="return_b", data=shares)
ax1.set(xlabel=' , %', ylabel=' B, %')
plt.show()
ax2 = sns.regplot(x="return_a", y="return_c", data=shares)
ax2.set(xlabel=' , %', ylabel=' C, %')
plt.show()
ax3 = sns.regplot(x="return_b", y="return_c", data=shares)
ax3.set(xlabel=' B, %', ylabel=' C, %')
plt.show()
```
