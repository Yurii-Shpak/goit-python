```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
from scipy.interpolate import interp1d
from scipy import integrate

speed = [25, 35, 45, 30, 60, 120, 100, 100, 70, 75, 80, 65]
time = np.linspace(0, 11, 12, dtype='int')
print(time)

plt.plot(time, speed, 'o')
plt.xlim(0,11)
plt.xticks(time)
plt.xlabel('Time, h')
plt.ylim(0,130)
plt.yticks(range(0, 131, 10))
plt.ylabel('Speed, km/h')
plt.grid(alpha=0.3)
f = interp1d(time, speed, kind='cubic')
interp_time = np.linspace(0, 11, 10000)
plt.plot(interp_time, f(interp_time))
plt.show()
print(f'Інтерполяція "cubic".\nІнтеграл на проміжку [0, 11] = {round(integrate.quad(f, 0, 11)[0], 2)}')
```
