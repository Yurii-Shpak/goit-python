## Медианная цена? Ответ: <b>$11</b>

```python
max_rating = books[['user_rating']].max()
max_rating
```

## Какой рейтинг в датасете наивысший? Ответ: <b>4.9</b>

```python
books[books[['user_rating']] == max_rating]['user_rating'].count()
```

## Сколько книг имеют такой рейтинг? Ответ: <b>52</b>

```python
books[books['reviews'] == books['reviews'].max()][['name', 'author', 'reviews']]
```

## У какой книги больше всего отзывов? Ответ: <b>"Where the Crawdads Sing" (Delia Owens)</b>

```python
books_2015 = books[books['year'] == 2015]

pd.options.display.max_colwidth = 100
books_2015[books_2015['price'] == books_2015['price'].max()][['name', 'author', 'price']]
```

## Из тех книг, что попали в Топ-50 в 2015 году, какая книга самая дорогая? Ответ: <b>"Publication Manual of the American Psychological Association", 6th Edition (American Psychological Association)</b>

```python
books[(books['year'] == 2010) & (books['genre'] == 'Fiction')].count()[0]
```

## Сколько книг жанра Fiction попали в Топ-50 в 2010 году? Ответ: <b>20</b>

```python
books[(books['year'].isin([2010, 2011])) & (books['user_rating'] == 4.9)].count()[0]
```

## Сколько книг c рейтингом 4.9 попало в рейтинг в 2010 и 2011 годах? Ответ: <b>1</b>

```python
books[(books['year'] == 2015) & (books['price'] < 8)].sort_values('price')
```

## Какая книга последняя в отсортированном списке? Ответ: <b>"Old School (Diary of a Wimpy Kid #10)" (Jeff Kinney)</b>

```python
books[['genre', 'price']].groupby('genre').agg(['min','max'])
```

## Максимальная цена для жанра Fiction: Ответ: <b>$82</b>

## Минимальная цена для жанра Fiction: Ответ: <b>$0</b>

## Максимальная цена для жанра Non Fiction: Ответ: <b>$105</b>

## Минимальная цена для жанра Non Fiction: Ответ: <b>$0</b>

```python
count_by_authors = books.groupby('author')['author'].count().to_frame().rename(columns={'author': 'count'}).sort_values('count')
count_by_authors
```

## Какой размерности вышла таблица? Ответ: <b>248 строк</b>

## У какого автора больше всего книг? Ответ: <b>Jeff Kinney</b>

## Сколько книг у этого автора? Ответ: <b>12</b>

```python
mean_rating_by_authors = books.groupby('author')['user_rating'].agg('mean').round(2).to_frame().rename(columns={'user_rating': 'mean_rating'}).sort_values('mean_rating')
mean_rating_by_authors
```

## У какого автора средний рейтинг минимальный? Ответ: <b>Donna Tartt</b>

## Какой у этого автора средний рейтинг? Ответ: <b>3.9</b>

```python
united_df = pd.concat([count_by_authors, mean_rating_by_authors], axis=1).sort_values(['count', 'mean_rating'])
united_df
```
