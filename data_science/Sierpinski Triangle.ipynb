```python
import matplotlib.pyplot as plt
from random import randint

x1 = -10
y1 = 0
x2 = 10
y2 = 0
x3 = 0
y3 = 10
x0 = 0
y0 = 0

x_coords = []
y_coords = []

for i in range(0,100000):
    r = randint(1,3)
    if r == 1:
        x0 = (x0 + x1) / 2
        y0 = (y0 + y1) / 2
    if r == 2:
        x0 = (x0 + x2) / 2
        y0 = (y0 + y2) / 2
    if r == 3:
        x0 = (x0 + x3) / 2
        y0 = (y0 + y3) / 2
    x_coords.append(x0)
    y_coords.append(y0)

plt.title(' ')
plt.scatter(x_coords, y_coords, s=0.1)
plt.show()
```
