```python
import pandas as pd
import numpy as np

tables = pd.read_html("https://uk.wikipedia.org/wiki/%D0%9D%D0%B0%D1%81%D0%B5%D0%BB%D0%B5%D0%BD%D0%BD%D1%8F_%D0%A3%D0%BA%D1%80%D0%B0%D1%97%D0%BD%D0%B8#%D0%9D%D0%B0%D1%80%D0%BE%D0%B4%D0%B6%D1%83%D0%B2%D0%B0%D0%BD%D1%96%D1%81%D1%82%D1%8C")
table = tables[10]
table
```

# Замените в таблице значения "—" на значения NaN

```python
table.replace('—', np.nan, inplace=True)
table
```

```python
table.dtypes
```

# Замените типы не числовых колонок на числовые.

```python
table.fillna(0, inplace=True)
table[['1950', '1960', '1970', '1990', '2000', '2012', '2014', '2019']] = table[['1950', '1960', '1970', '1990', '2000', '2012', '2014', '2019']].astype(float)
print(table)
table.dtypes
```

### Корректируем ошибку в данных. Необходимо все числовые значення разделить на 10

```python
table[['1950', '1960', '1970', '1990', '2000', '2012', '2014', '2019']] = table[['1950', '1960', '1970', '1990', '2000', '2012', '2014', '2019']].apply(lambda x: x/10)
print(table)
```

```python
table.replace(0, np.nan, inplace=True)
print(table)
```

# Посчитайте, какая доля пропусков содержится в каждой колонке

```python
table.isnull().sum()
```

# Удалите из таблицы данные по всей стране

```python
table.drop(27, inplace=True)
table
```

# Замените отсутствующие данные в столбцах средними значениями по этим столбцам

```python
table.fillna(value=table.mean(numeric_only=True).round(1), inplace=True)
table
```

# Получите список регионов, где уровень рождаемости в 2019 году был выше среднего по Украине

```python
table.loc[table['2019'] > table[['2019']].mean()[0], ['Регіон', '2019']]
```

В каком регионе была самая высокая рождаемость в 2014 году?

```python
table.loc[table['2014'] == table['2014'].max(), ['Регіон', '2014']]
```

# Постройте столбцовую диаграмму рождаемости по регионам в 2019 году

```python
import matplotlib.pyplot as plt

table[['Регіон', '2019']].plot(kind='barh', x='Регіон', legend=False, ylabel='', xticks=range(12), grid=True)
plt.show()
```
